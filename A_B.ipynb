It seems like the problme is that the experiment is not randomized across age groups and gender. Meaning that the distribution of age and sex are different for the experimental group and the control group. Mexico seems to be balanced in gender distribution, but not age. From Argentina and El Salvador, we can see that the gender and age distribution are both biased, which could introduced biased in the experiment.

A pipeline should check the distribution of age, gender and where they are from for experimental and control group should be designed - we should make sure that the P(T=1|X) = P(T=0|X)
